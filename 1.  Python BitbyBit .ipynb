In Python, variable names must adhere to certain rules and conventions. Here are the rules for creating valid Python variable names:

> **Valid Characters:** Variable names can only contain letters (both uppercase and lowercase), digits, and underscores (). They must start with a letter (a-z, A-Z) or an underscore ().

>**Case Sensitivity:** Python is case-sensitive, meaning myVariable and myvariable are considered different variables.

>**No Spaces or Special Characters:** Variable names cannot contain spaces or special characters like !, @, #, $, %, etc.

> You cannot use Python's reserved keywords as variable names. These are words that Python has reserved for specific purposes, such as if, else, while, for, def, class, import, etc.

>**Meaningful Names:** Variable names should be meaningful and descriptive. Use names that indicate the purpose or content of the variable.

> Variable names cannot start with a digit (0-9).

>**No Punctuation:** Avoid using punctuation marks like hyphens (-), periods (.), or slashes (/) in variable names.

> While there's no strict limit, it's recommended to keep variable names reasonably short and descriptive. Long variable names can make your code harder to read.

PEP 8 Guidelines: Adhere to the PEP 8 style guide for Python code, which recommends using lowercase letters for variable names, separating words with underscores (snake_case), and using uppercase for constants (ALL_CAPS).

Not all primitive data types in Python are immutable. Immutable means that once an object of that type is created, its value cannot be changed. Here's a breakdown of the mutability of common primitive data types in Python:

**Immutable Primitive Data Types:**

int: Integer values are immutable. When you perform operations on an integer, you create a new integer object rather than modifying the original one.
float: Floating-point numbers are also immutable, for the same reason as integers.
str: Strings are immutable. Any operation that seems to modify a string actually creates a new string object.
bool: Booleans are immutable. You can't change a True to a False or vice versa; you create a new boolean object.

**Mutable Primitive Data Type:**

bytes and bytearray: These are sequences of bytes that represent raw binary data. **bytes** objects are immutable, while **bytearray objects** are mutable.
So, while most primitive data types are immutable, there are exceptions like bytearray that are mutable

### **Size and type of data types**
### example that demonstrates how to check the type and size of basic data types in Python:


> In this code, we declare variables x, y, z, and p, and demonstrate how to check their type and size using built-in functions and methods.

```python
# Checking the type and size of basic data types
x = 10
print(type(x))  # Output: <class 'int'>
print(x.bit_length())  # Output: 4

y = 3.14
print(type(y))  # Output: <class 'float'>
print(y.hex())  # Output: 0x1.91eb851eb851fp+1

z = "Hello, World!"
print(type(z))  # Output: <class 'str'>
print(len(z))  # Output: 13

p = True
print(type(p))  # Output: <class 'bool'>
print(p.__sizeof__())  # Output: 28
```

The __sizeof__() method returns the size of the object in bytes. However, the exact size of objects can be influenced by various factors, including the underlying memory management and alignment mechanisms of the Python interpreter and the specific system architecture.

On a typical 64-bit system, the size of a bool object is 28 bytes because it includes not only the space for the actual boolean value (True or False) but also additional memory overhead due to the Python object infrastructure and alignment requirements. This overhead includes information like the object's reference count and its type.

Please note that the exact size can vary between different Python implementations (such as CPython, Jython, IronPython, etc.) and different versions of Python. The line of code **print(f"{a} == {b} ? {a == b}")** is an example of a formatted string in Python.

In a formatted string, any expressions **inside curly braces ({}) are evaluated and their values are inserted into the string.** In this case, the expressions are a, b, and a == b.

The == operator is a comparison operator that tests whether two values are equal. In this line of code, we're testing whether a is equal to b, and the result of the test (True or False) is inserted into the formatted string.

So, the output of this line of code depends on the values of a and b. If a is equal to b, the output will be something like 5 == 5 ? True. If a is not equal to b, the output will be something like 5 == 10 ? False.

## Break and Continue

The continue and break keywords are control flow statements in Python that are used to modify the behavior of loops (both for and while loops). They allow you to control the execution of loops based on certain conditions.


**break Keyword:**
The break keyword is used to immediately exit the loop, regardless of whether the loop's condition has been satisfied or not. When break is encountered inside a loop, the loop is terminated, and the program continues with the next statement after the loop.

```python
for i in range(1, 6):
    if i == 3:
        break
        print(i)
```

```python
start = 1
end = 10

for num in range(start, end + 1):
    if num % 2 == 0:
        print(f"The first even number in the range is: {num}")
        break
```

In this example, the loop iterates through the numbers from start to end (inclusive). The if condition checks if the current number is even. Once an even number is found, the break statement is encountered, which immediately exits the loop. The program then prints the first even number it found and terminates.

**continue Keyword:**
The continue keyword is used to skip the current iteration of the loop and move to the next iteration. When continue is encountered inside a loop, the rest of the loop's body for the current iteration is skipped, and the loop proceeds to the next iteration.

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

here's an example of using the break statement within an if-elif-else structure to create a simple calculator program that performs basic arithmetic operations. The program will repeatedly prompt the user for input until they choose to exit by typing "exit".



In this example, the program presents a simple calculator menu to the user. If the user selects a valid operation (1 to 4), the program performs the chosen operation and displays the result. If the user selects "exit," the program breaks out of the loop and terminates. If the user enters an invalid choice, the program informs them and continues to the next iteration of the loop.

The break statement is used to exit the loop when the user chooses to quit by typing "exit." The continue statement is used within the division calculation to skip that iteration if the user attempts to divide by zero, preventing a ZeroDivisionError.


```python
while True:
    print("Choose an operation:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("Type 'exit' to quit")

    choice = input("Enter your choice: ")

    if choice == "exit":
        print("Exiting calculator.")
        break
    elif choice in ("1", "2", "3", "4"):
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == "1":
            result = num1 + num2
        elif choice == "2":
            result = num1 - num2
        elif choice == "3":
            result = num1 * num2
        elif choice == "4":
            if num2 == 0:
                print("Cannot divide by zero!")
                continue  # Skip the current iteration
            result = num1 / num2

        print("Result:", result)
    else:
        print("Invalid choice. Please enter a valid option.")
```

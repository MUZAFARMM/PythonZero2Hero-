## 5. Formal and Actual Arguments

1. Formal arguments are defined in the function's parameter list.
2. Actual arguments are the values passed when calling the function.
These can be futher divided into four types:
*   Positional Arguments
*   Keyword Arguments
*   Default Arguments
*   Variable Length Arguments



**Positional Arguments**

Positional arguments are the most common type of arguments in Python. They are passed to a function in the order they appear in the function's parameter list. The number and order of arguments must match the function's parameter list.


```python
# Positional Arguments
def add(a, b):
    result = a + b
    return result

sum_result = add(5, 3)  # 5 is assigned to 'a', and 3 is assigned to 'b'
print("Sum:", sum_result)  # Output: Sum: 8
```

**Keyword Arguments**

Keyword arguments are passed to a function using the parameter names as keywords, allowing you to specify the values for specific parameters regardless of their order in the parameter list.

```python
# Keyword Arguments
def greet(name, message):
    return f"Hello, {name}! {message}"

greeting = greet(message="How are you?", name="Alice")
print(greeting)  # Output: Hello, Alice! How are you?
```

**Default Arguments:**

Default arguments have default values specified in the function definition. If a value is not provided for a default argument when the function is called, the default value is used.



```python
# Default Arguments
def greet(name, message="Hello"):
    return f"{message}, {name}!"

greeting1 = greet("Alice")  # No message provided, so the default "Hello" is used.
greeting2 = greet("Bob", "Hi")  # "Hi" is provided for the message.

print(greeting1)  # Output: Hello, Alice!
print(greeting2)  # Output: Hi, Bob!
```

**Variable Length Arguments:**

Python allows you to create functions that accept a variable number of arguments. The *args and *kwargs syntax is used for this purpose.

```python
# Variable Length Arguments
# Example with *args:
def sum_values(*args):
    print(args)
    print(type(args))# this is a tuple
    total = 0
    for num in args:
        total += num
    return total

result = sum_values(1, 2, 3, 4, 5)
print("Sum:", result)  # Output: Sum: 15

# In this example, *args allows the sum_values function to accept any number of positional arguments, and it calculates their sum.
```

```python
# Example with **kwargs:
def display_info(**kwargs):
    print(type(kwargs)) # so the kwargs is a dictionary
    for key, value in kwargs.items():
        print(f"{key}: {value}")

display_info(name="Alice", age=30, city="New York")



# In this example, **kwargs allows the display_info function to accept a variable number of keyword arguments, which are then printed.

# These different types of function arguments provide flexibility when defining and calling functions, making your code more versatile and readable.
```

## 6. Local and Global Arguments

In Python, variables can be categorized into two main types: local variables and global variables. Their scope determines where they can be accessed and modified in a program.



**1. Local Variables:**

Local variables are defined within the scope of a function.
They are only accessible within the function in which they are defined.
They are temporary and exist only while the function is executing.
Local variables cannot be directly accessed or modified from outside the function.
Example of Local Variables:

```python
def my_function():
    # This 'x' is a local variable
    x = 10
    print("Inside the function:", x)

my_function()
# Trying to access 'x' outside the function will result in an error
print("Outside the function:", x)  # Uncommenting this line will raise an error
```

In this example, x is a local variable, and it is only accessible within the my_function().

**2. Global Variables:**

Global variables are defined at the outermost scope of a program or in a module.
They are accessible throughout the program, including within functions.
Global variables have a longer lifespan and persist beyond the function that created them.
They can be accessed and modified from any part of the program, including within functions.

Example of Global Variables:

## 7 Recursive Functions

Recursion is a programming technique in which a function calls itself to solve a problem. In other words, a recursive function is a function that solves a problem by breaking it down into smaller instances of the same problem. These smaller instances are typically simpler or smaller in size, and the function continues to call itself until it reaches a base case, which is a condition that indicates when the recursion should stop.

Key characteristics of recursion:

**Base Case:** Every recursive function must have one or more base cases. A base case is a condition or set of conditions that specify when the recursion should terminate. Without a base case, the recursive function can run indefinitely, leading to a stack overflow error.

**Recursive Case:** The recursive function must have one or more cases in which it calls itself with modified input data, moving closer to the base case. These recursive calls should eventually lead to the base case.

**Function Decomposition:** The problem is broken down into smaller, more manageable subproblems that are solved independently by recursive calls.

Recursion is particularly useful for solving problems that can be naturally divided into smaller, similar subproblems, such as calculating factorials, traversing tree-like data structures (e.g., binary trees), or solving problems involving sequences (e.g., Fibonacci sequence).

Here's an example of a recursive function to calculate the factorial of a number in Python:



image.png

```python
def factorial(n):
    if n == 0:  # Base case
        return 1
    else:
        return n * factorial(n - 1)  # Recursive case

result = factorial(5)  # Calculates 5!
print("Factorial of 5:", result)  # Output: Factorial of 5: 120
```

## 8 Function Decorators
A decorator function in Python is a higher-order function that takes another function as an argument and extends or modifies the behavior of the target function without permanently modifying its source code. Decorators are often used to add functionality to functions or methods, such as logging, access control, memoization, or timing.

In Python, decorators are typically implemented as functions or classes. Here's a simple example of a decorator function:

```python
# Decorator function
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

# Function to be decorated
def say_hello():
    print("Hello!")

# Decorate the function
decorated_say_hello = my_decorator(say_hello)

# Call the decorated function
decorated_say_hello()
```

In this example:

1. my_decorator is a decorator function that takes another function func as its argument.
2. Inside my_decorator, a new inner function wrapper is defined. This inner function can extend or modify the behavior of func.
3. wrapper includes code that runs before and after calling func. This allows you to add functionality before and after the original function is executed.

To apply the decorator to a function, you simply use the @ symbol followed by the decorator function's name above the target function definition:

```python
@my_decorator
def say_hello():
    print("Hello!")

# Call the decorated function
say_hello()
```

When you call say_hello(), it is actually the decorated version of say_hello that gets executed, and it produces the following output:

Decorators are a powerful feature in Python, commonly used in web frameworks like Flask and Django for routing, authentication, and other middleware tasks. They help keep code modular, clean, and easy to read by separating concerns and allowing you to add or modify behavior without altering the original function's code.

## 9. Generator Function
A generator function in Python is a special type of function that allows you to create iterators (iterable sequences of values) in a more memory-efficient and lazy manner.

Generator functions are defined using the yield keyword instead of return. When a generator function is called, it doesn't execute the function's body immediately. Instead, it returns a generator object that can be used to iterate over the values produced by the generator function on-the-fly.

Here's a basic example of a generator function:

```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

# Create a generator object
counter = countdown(5)

# Iterate over the values produced by the generator
for value in counter:
    print(value)
```

In this example:

1. The countdown function is a generator function that takes an integer n as its argument.
2. Inside the function, there's a while loop that counts down from n to 1.
3. Instead of using return, the yield keyword is used to yield each value one by one. When yield is encountered, the current state of the function is saved, and the yielded value is returned to the caller.

4. The function execution is paused at the yield statement, and it can be resumed later from where it left off when the generator is iterated.
When you create a generator object (counter) by calling countdown(5), the function is not executed immediately. Instead, it returns a generator object. You can then iterate over this generator object using a for loop, as shown in the example. The generator function is executed incrementally as you iterate over it, producing values one at a time.

Generator functions are particularly useful when working with large datasets or when you want to generate values lazily without precomputing and storing them in memory. They can also be used to represent infinite sequences, such as streaming data or mathematical sequences, without the need for infinite memory.

## 10. Anonymous Functions or Lambdas


In Python, a lambda function, also known as an anonymous function or a lambda expression, is a small, unnamed function that is defined using the lambda keyword. Lambda functions are often used for short, simple operations where a full function definition is not necessary. They are a way to create small, throwaway functions without having to formally define them using the def keyword.

The basic syntax of a lambda function is as follows:

**lambda arguments: expression**

Arguments are the input parameters to the function.
expression is a single expression that is evaluated and returned as the result of the lambda function. Here's a simple example of a lambda function that adds two numbers:

```python
add = lambda x, y: x + y
result = add(5, 3)
print(result)  # Output: 8
# In this example, the lambda function takes two arguments x and y and returns their sum.
```

Lambda functions are often used in situations where a small, anonymous function is needed for a short duration, such as when sorting or filtering data using higher-order functions like map(), filter(), or sorted(). For example:

```python
# Using lambda with the sorted() function to sort a list of tuples by the second element
data = [(1, 5), (3, 2), (2, 8)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)  # Output: [(3, 2), (1, 5), (2, 8)]
```

### Using map() with lambda to square a list of numbers:
The map() function applies a given function to each item in an iterable (e.g., a list) and returns an iterable with the results.

It takes two arguments: the function to apply and the iterable.

The result is an iterator, so you often convert it to a list or another iterable to see the results.

In this example, the map() function applies the lambda function to each element in the numbers list, squaring each number.

```python
numbers = [1, 2, 3, 4, 5]

# Using map() to square each number in the list
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
```

### Using filter() with lambda to filter even numbers from a list and square them:

The filter() function filters items from an iterable based on a given condition (a function that returns True or False).

It takes two arguments: the filtering function and the iterable.

It returns an iterator with the items that meet the specified condition.

In this example, the filter() function is used with a lambda function to select even numbers from the numbers list, and then the map() function is used to square each of the selected even numbers.

```python
numbers = [1, 2, 3, 4, 5, 6]
f=filter(lambda x: x % 2 == 0, numbers)
print(f)
print(list(f))
# Using filter() and lambda to select even numbers, then mapping to square them
even_squared = list(map(lambda x: x ** 2, filter(lambda x: x % 2 == 0, numbers)))
print(even_squared)  # Output: [4, 16, 36]
```

### Using reduce() with lambda to find the product of squared numbers in a list:

The reduce() function performs a cumulative operation on the items of an iterable, reducing it to a single value.

It takes two arguments: a binary function (a function that takes two arguments) and the iterable.

The function is applied cumulatively to the items from left to right.

It's important to note that reduce() needs to be imported from the functools module in Python 3.

In this example, reduce() is used with a lambda function to calculate the product of the squared numbers in the list.


```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]
m=map(lambda x: x ** 2, numbers)
print(list(m))
# print(m)

# Using reduce() and lambda to find the product of squared numbers
# product_of_squares = reduce(lambda x, y: x * y, m)
product_of_squares = reduce(lambda x, y: x * y, list(m))
print(m)
print(product_of_squares)  # Output: 14400
# print(list(m))
```

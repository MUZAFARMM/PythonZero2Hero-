## **Opening and Closing Files-**

* To perform file operation ,it must be opened first then after reading ,writing, editing operation can be performed.

* To create any new file  then too it must be opened.

* On opening of any file ,a file relevant  structure is created in memory as well as memory space is created to  store contents.

### **Absolute Path vs Relative Path**

One must be familiar with absolute & relative path before starting file related operations.

The absolute path is the full path to some place on your computer.

The relative path is the path  to some file with respect to your current working directory (PWD).

For example: Absolute path: C:/users/admin/docs/staff.txt

If PWD is C:/users/admin/, then the relative path to staff.txt would be: docs/staff.txt

Note, **PWD + relative path = absolute path**. Now, if we write some scripts which check if a file exists.

os.chdir("C:/users/admin/docs")  
os.path.exists("staff.txt")

This returns TRUE if stuff.txt exists and it works.  

Now, instead if we write,  

os.path.exists("C:/users/admin/docs/staff.txt")  

This will returns TRUE.

If we don't know where the user executing the script from, it is best to compute the absolute  path on the user's system using os and 	file 	.

file 	is a global variable set on every Python  script that returns the relative path to the *.py file  that contains it.

````python
import os

# Get the current working directory
print("Current working directory:", os.getcwd())

# Create a directory named "newdir1"
os.mkdir("newdir1")

# Change the current working directory to "newdir1"
os.chdir("newdir1")
print("Current working directory after changing:", os.getcwd())

# Create a text file in the "newdir1" directory
file_path = "sample.txt"
with open(file_path, "w") as file:
    file.write("Hello, this is a sample text file!\n")
    file.write("Adding some more text here.\n")

# Display the absolute directory path
my_absolute_dirpath = os.path.abspath(os.getcwd())
print("Absolute directory path:", my_absolute_dirpath)

# Read and display the content of the text file
with open(file_path, "r") as file:
    file_content = file.read()
    print("File content:")
    print(file_content)
````

### Getting & Resetting the Files Position

The tell() method of python tells us the current position within the file,where as The seek(offset[,  from]) method changes the current file position.

If from is 0, the beginning of the file to seek. If it is set  to 1, the current position is used . If it is set to 2 then the end of the file would be taken as seek  position. The offset argument indicates the number of bytes to be moved.

````python
f = open("a.txt", 'w')
line = 'Welcome to python.mykvs.in\nRegularly visit python.mykvs.in'
f.write(line)
f.close()
````

````python
f = open("a.txt", 'rb+')
print(f.tell())
print(f.read(7)) # read seven characters
print(f.tell())
print(f.read())
print(f.tell())
f.seek(9,0) # moves to 9 position from begining
print(f.read(5))
f.seek(4, 1) # moves to 4 position from current location
print(f.read(5))
f.seek(-5, 2) # Go to the 5th byte before the end  print(f.read(5))
````

## **Modify a Text file**

There is no way to insert into the middle of a file without re-writing  it. We can append to a file or overwrite part of it using seek but if  we want to add text at the beginning or the middle, we'll have to  rewrite it.

What we usually do for modification is read from the file, make the  modifications and write it out to a new file called temp.txt or  something like that.

This is better than reading the whole file into  memory because the file may be too large for that. Once the  temporary file is completed, rename it the same as the original file.

This is a good, safe way to do it because if the file write crashes or  aborts for any reason in between, we still have our untouched

### Replace string in the same File

What have we done here? Open file dummy.txt in read text mode fin.read() reads whole text in dummy.txt to the variable data.

data.replace() replaces all the occurrences of my with your in the whole text.  

fin.close() closes the input file dummy.txt. In the last three lines, we are opening dummy.txt in write text  mode and writing the  data to dummy.txt in replace mode.

Finally closing the file dummy.txt.

````python
fin = open("sample.txt", "r")  # reading file in text mode
data = fin.read()
data = data.replace('my', 'your')
fin.close()
fin = open("dummy.txt", 'w')
fin.write(data)
fin.close()
````

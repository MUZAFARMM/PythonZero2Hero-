###  Lists



* In Python, a list is a collection of values that can be of different types, such as integers, floats, strings, and other objects.
* Lists are ordered, meaning that the order in which items are added to the list is preserved.
* Lists are also mutable, meaning that they can be changed after they are created by adding, removing, or changing elements.
* Lists are created using square brackets [], with each element separated by a comma. For example, the following code creates a list of integers:

```python
my_list = [1, 2, 3, 4, 5]
type([])
```

* We can access individual elements of a list using their index, which starts at 0. For example, **my_list[0]** returns the first element of the list, which is 1.

* Lists have a number of useful methods and functions that can be used to manipulate and work with them. Some of the most commonly used methods and functions for lists include:

* append(): adds an element to the end of the list.
* extend(): adds the elements of another list to the end of the list.
* insert(): inserts an element at a specific index in the list.
* remove(): removes the first occurrence of an element from the list.
* pop(): removes and returns the last element of the list.
* index(): returns the index of the first occurrence of an element in the list.
* count(): returns the number of times an element appears in the list.
* sort(): sorts the elements of the list in ascending order.
* reverse(): reverses the order of the elements in the list.
Here is an example of creating a list, accessing elements, and using some of the methods:

* Lists can also be sliced to extract a subset of the items in the list.
* The syntax for slicing is **my_list[start:stop:step]**, where start is the index of the first item to include, stop is the index of the first item to exclude, and step is the interval between items. If any of these values are omitted, they default to the beginning, end, and step size of 1, respectively.
* Python also provides a variety of built-in functions for working with lists, such as len() to get the length of a list, sorted() to sort a list in ascending order, and sum() to get the sum of all the elements in a list of numbers.

* Slicing in Python is a way to extract a specific section of a string, list, or any sequence of data.
* Slicing a list in Python means that a part of the list is taken out, and this part can be either a single element or a group of elements.

The syntax for slicing a list in Python is as follows:

* my_list[start:stop:step]
  where:

* start is the starting index of the slice (inclusive)
* stop is the ending index of the slice (exclusive)
* step is the number of elements to skip between each element of the slice (optional, defaults to 1)
Here are a few examples to demonstrate slicing in lists:

```python
my_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# slice from index 1 (inclusive) to 4 (exclusive)
slice1 = my_list[1:4]
print(slice1)  # Output: [2, 3, 4]

# slice from index 2 (inclusive) to the end of the list
slice2 = my_list[2:]
print(slice2)  # Output: [3, 4, 5, 6, 7, 8, 9]

# slice from the beginning of the list to index 5 (exclusive)
slice3 = my_list[:5]
print(slice3)  # Output: [1, 2, 3, 4, 5]

# slice from index 1 to index 8, skipping every other element
slice4 = my_list[1:8:2]
print(slice4)  # Output: [2, 4, 6, 8]
```

* In the first example, we sliced the list to include only the elements from index 1 to index 4 (exclusive).
* In the second example, we sliced the list to include all elements from index 2 (inclusive) to the end of the list.
* In the third example, we sliced the list to include all elements from the beginning of the list to index 5 (exclusive).
* In the fourth example, we sliced the list to include only the elements from index 1 to index 8 (exclusive), but we skipped every other element using a step of 2.

Slicing can also be used to modify a list. For example, the following code replaces the second and third elements of the list with the numbers 10 and 11:

```python
my_list = [1, 2, 3, 4, 5]
my_list[1:3] = [10, 11]
print(my_list)  # Output: [1, 10, 11, 4, 5]
```

In this example, we used slicing to target the second and third elements of the list (using the indexes 1:3), and we assigned the list [10, 11] to this slice. This modified the original list to become [1, 10, 11, 4, 5].

## Sets



* In Python, a set is an unordered collection of unique elements, meaning that each element can only occur once in the set.
* It is similar to a list or tuple, but with two main differences:
** first, sets are unordered, meaning the order in which elements are added to the set is not preserved; and
** second, sets can only contain unique elements, so duplicates are automatically removed.

* Sets are implemented in Python as a built-in class, and can be created using curly braces {} or the set() constructor.
* Elements in a set can be of any data type that is hashable, meaning that it has a unique hash value that remains constant throughout its lifetime. Common hashable types in Python include numbers, strings, and tuples, but lists and dictionaries are not hashable.

Here are some examples of how to create sets in Python:

A Set is an unordered collection data type that is iterable, mutable, and has no duplicate elements. Python's set class represents the mathematical notion of a set.This is based on a data structure known as a hash table

```python
# Create a set using curly braces
my_set = {1, 2, 3, 4, 5}

# Create a set using the set() constructor
my_set = set([1, 2, 3, 4, 4, 5])
my_set
# Create an empty set using the set() constructor
# my_set = set()

```

* Sets support a variety of operations, including
** union,
** intersection,
** difference, and
** symmetric difference.

 These can be performed using built-in methods or operators, as shown in the following examples:

```python
# Create two sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Union of two sets
print(set1.union(set2))
print(set1 | set2)

# Intersection of two sets
print(set1.intersection(set2))
print(set1 & set2)

# Difference of two sets
print(set1.difference(set2))
print(set1 - set2)

# Symmetric difference of two sets
print(set1.symmetric_difference(set2))
print(set1 ^ set2)
```

```python
# create sets
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 7}
set3 = {2, 6, 8}

# check if sets are disjoint
disjoint = set1.isdisjoint(set2)
print("Set1 and Set2 are disjoint:", disjoint)

# check if a set is superset of another set
superset = set1.issuperset(set2)
print("Set1 is superset of Set2:", superset)

# check if a set is subset of another set
subset = set1.issubset(set2)
print("Set1 is subset of Set2:", subset)

# check if a set is superset of another set
superset = set1.issuperset(set3)
print("Set1 is superset of Set3:", superset)

# check if a set is subset of another set
subset = set3.issubset(set1)
print("Set3 is subset of Set1:", subset)
```

In the example, set1, set2, and set3 are three sets that we have defined. We then use the **isdisjoint()**, **issuperset()**, and **issubset()** methods to check the relationship between the sets.

* **isdisjoint()** method checks whether two sets are disjoint, i.e., whether they have no common elements. In the code, we use **isdisjoint()** method to check whether set1 and set2 are disjoint.

* **issuperset()** method checks whether one set is a superset of another set. A set is said to be a superset of another set if it contains all the elements of the other set. In the code, we use **issuperset()** method to check whether set1 is a superset of set2 and set3.

* **issubset()** method checks whether one set is a subset of another set. A set is said to be a subset of another set if all its elements are present in the other set. In the code, we use **issubset()** method to check whether set1 is a subset of set2 and set3.

## Dictionaries



* In Python, a dictionary is a collection of unordered, changeable, and indexed key-value pairs.
* It is also known as an associative array, map, or hash table. Dictionaries are declared with curly braces {} and each item in the dictionary is separated by a comma.
* The key and value are separated by a colon ":". Keys in the dictionary must be unique and can be of any immutable type such as a string, integer or tuple.

Here's an example of a dictionary in Python:

```python
# Creating a dictionary
student = {"name": "John", "age": 25, "course": "Python"}

# Accessing a dictionary
print(student["name"])
print(student["age"])
print(student.get("age"))

# Changing the value of a key
student["course"] = "Java"
print(student)

# Adding a new key-value pair
student["gender"] = "Male"
print(student)

# Removing a key-value pair
del student["age"]
print(student)
```

```python
print(student.items())
```

```python
# Looping through a dictionary
for key, value in student.items():
    print(key, ":", value)

# Checking if a key exists in a dictionary
if "name" in student:
    print("Name is present in the dictionary")
```

There are several built-in functions that can be used with Python dictionaries:
* len() - returns the number of key-value pairs in the dictionary
* dict() - creates a new dictionary
* keys() - returns a view object of the keys in the dictionary
* values() - returns a view object of the values in the dictionary
* items() - returns a view object of the key-value pairs in the dictionary
* get() - returns the value of the specified key. If the key does not exist, it * returns the default value specified in the function (or None if not specified)
* pop() - removes and returns the value of the specified key
* popitem() - removes and returns a random key-value pair from the dictionary
* clear() - removes all key-value pairs from the dictionary
* update() - updates the dictionary with the key-value pairs from another dictionary or an iterable of key-value pairs

```python
# Create a dictionary
my_dict = {'apple': 3, 'banana': 5, 'orange': 2}

# Get the length of the dictionary
print(len(my_dict))  # Output: 3

# Create a new dictionary
new_dict = dict([('grape', 4), ('pineapple', 1)])
print(new_dict)  # Output: {'grape': 4, 'pineapple': 1}

# Get a view object of the keys in the dictionary
keys_view = my_dict.keys()
print(keys_view)  # Output: dict_keys(['apple', 'banana', 'orange'])

# Get a view object of the values in the dictionary
values_view = my_dict.values()
print(values_view)  # Output: dict_values([3, 5, 2])

# Get a view object of the key-value pairs in the dictionary
items_view = my_dict.items()
print(items_view)  # Output: dict_items([('apple', 3), ('banana', 5), ('orange', 2)])

# Get the value of a specific key
apple_value = my_dict.get('apple')
print(apple_value)  # Output: 3

# Remove and return the value of a specific key
banana_value = my_dict.pop('banana')
print(banana_value)  # Output: 5
print(my_dict)
# Remove and return a random key-value pair
random_pair = my_dict.popitem()
print(random_pair)  # Output: ('orange', 2)

# Remove all key-value pairs from the dictionary
my_dict.clear()
print(my_dict)  # Output: {}

# Update the dictionary with key-value pairs from another dictionary
my_dict.update(new_dict)
print(my_dict)  # Output: {'grape': 4, 'pineapple': 1}
```

## Tuples


In Python, a tuple is an immutable sequence of values, similar to a list. However, unlike a list, a tuple cannot be modified once it is created. Tuples are often used to group related data together.

Tuples are represented with parentheses, and the elements are separated by commas. For example:

```python
my_tuple = (1, 2, 3, "hello", 4.5)
```

Tuples can also be created without parentheses, by separating the elements with commas. This is called tuple packing. For example:

```python
my_tuple = 1, 2, 3, "hello", 4.5
t1=()
type(my_tuple)
type(t1)
type((4))
```

To access the elements of a tuple, you can use indexing, just like with a list. For example:

```python
my_tuple = (1, 2, 3, "hello", 4.5)
print(my_tuple[0])  # Output: 1
print(my_tuple[3])  # Output: "hello"
type(my_tuple [3])
```

Since tuples are immutable, you cannot modify their elements directly. However, you can create a new tuple that includes some or all of the elements of the original tuple. This is called tuple unpacking. For example:

```python
my_tuple = (1, 2, 3, "hello", 4.5)
a, b, c, d, e = my_tuple
# c, d, e = my_tuple
# unpack needs all place holders
print(a)  # Output: 1
print(d)  # Output: "hello"
```

Tuples support a variety of built-in functions, such as
* len()
* max()  
* min()
* count()
* index()
* They also support a variety of operators, such as +, *, and in.
* concatenation using "+"
* repeatition using " *"
* in
The "in " keyword in Python is used to test if a value is a member of a sequence or collection, and returns a boolean value. It can be used with many data types, including lists, tuples, strings, and sets.

```python
# Finding tmy_tuple = (1, 2, 3)
my_tuple = (1, 2, 3)
print(len(my_tuple))  # output: 3
print(max(my_tuple))
print(min(my_tuple))
```

```python
# Counting the number of occurrences of an element in a tuple:
my_tuple = (1, 2, 2, 3, 3, 3)
print(my_tuple.count(2))  # output: 2
print(my_tuple.count(3))  # output: 3
```

```python
# Finding the index of the first occurrence of an element in a tuple:
my_tuple = (1, 2, 3, 2, 3, 4, 5, 6)
print(my_tuple.index(2))  # output: 1
print(my_tuple.index(3))  # output: 2
```

```python
# Converting a tuple to a list:
my_tuple = (1, 2, 3)
my_list = list(my_tuple)
print(my_list)  # output: [1, 2, 3]
```

```python
# Converting a list to a tuple:
my_list = [1, 2, 3]
my_tuple = tuple(my_list)
print(my_tuple)  # output: (1, 2, 3)
```

# Queue using List
This program displays a menu of queue operations and takes input from the user at runtime. The queue list is used to store the elements in the queue. The while loop keeps running until the user chooses to exit the program. Inside the loop, the user's choice is obtained using the input() function, and the corresponding operation is performed using if-elif statements.

To take input from the user dynamically, you can use the input() function in Python. Here's an example:

```python
name = input("Enter your name: ")
age = input("Enter your age: ")

print("Your name is", name)
print("Your age is", age)
```

```python
queue = []

while True:
    print("Queue Operations:")
    print("1. Enqueue")
    print("2. Dequeue")
    print("3. Display Queue")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        item = input("Enter the item to be enqueued: ")
        queue.append(item)
        print(f"{item} enqueued to the queue")

    elif choice == 2:
        if not queue:
            print("Queue is empty")
        else:
            item = queue.pop(0)
            print(f"{item} dequeued from the queue")

    elif choice == 3:
        print("Current Queue:")
        print(queue)

    elif choice == 4:
        print("Exiting the program...")
        break

    else:
        print("Invalid choice. Please enter a valid choice!")



```

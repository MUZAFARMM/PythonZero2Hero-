**what happened internally when we create instance of a class?**

When you create an instance of a class in Python, several steps occur internally. Let's break it down step by step:

1. Memory Allocation: When you create an instance of a class, Python allocates memory to store the instance's attributes and methods.


2. __new__ Method (Optional): If the class has a __new__ method, it is called first. This method is responsible for creating and returning a new instance of the class. If you don't define __new__, the default __new__ from the base object class is called.

3. __init__ Method: After the instance is created, the __init__ method is called. This method initializes the attributes of the instance. It takes self (the instance being created) as its first parameter, followed by any additional parameters you specify.

3. Instance Binding: The instance (self) is bound to the class. This means that when you call a method or access an attribute on the instance, Python knows to associate it with that particular instance.

4. Return Instance: If the __new__ method returns an instance, that instance is returned. If __new__ is not defined or returns None, the newly created instance from step 2 is

````python
class MyClass:
    def __new__(cls, *args, **kwargs):
        instance = super().__new__(cls)
        print("Creating a new instance")
        return instance

    def __init__(self, name):
        self.name = name
        print("Initializing instance")

# Creating an instance of MyClass
obj = MyClass("Example")
````



Instance methods in Python are functions defined inside a class, and they operate on instances of that class. These methods take the instance as the first parameter (usually named self), allowing them to access and modify the instance's attributes. Instance methods are the most common type of methods in Python classes.



1.   Accessor Methods (Getters): These are methods that are used to access the value of private instance variables. They do not modify the instance's state. By convention, accessor methods start with "get."

**Example:**
```
class Student:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    def get_name(self):
        return self._name

    def get_age(self):
        return self._age

```
2.   Mutator Methods (Setters): These are methods that modify the value of private instance variables. They update the instance's state. By convention, mutator methods start with "set."

**Example:**
```
class Student:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    def set_name(self, new_name):
        self._name = new_name

    def set_age(self, new_age):
        self._age = new_age
```









````python
class Student:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    def get_name(self):
        return self._name

    def get_age(self):
        return self._age

    def set_name(self, new_name):
        self._name = new_name

    def set_age(self, new_age):
        self._age = new_age

# Creating instances of the Student class
student1 = Student("John Doe", 20)
student2 = Student("Jane Smith", 22)

# Accessing information using accessor methods
print(f"Student 1: Name - {student1.get_name()}, Age - {student1.get_age()}")
print(f"Student 2: Name - {student2.get_name()}, Age - {student2.get_age()}")

# Modifying information using mutator methods
student1.set_name("Johnny")
student1.set_age(21)

# Accessing modified information
print(f"Student 1 (after modification): Name - {student1.get_name()}, Age - {student1.get_age()}")
````

# **Instance Attributes VS Static Attributes**

Note: In the case of instance variables for every object a seperate copy will be created,but in the
case of static variables for total class only one copy will be created and shared by every object of
that class.

````python
class Test:
    x = 10

    def __init__(self):
        self.y = 20

# Create instances of the Test class
t1 = Test()
t2 = Test()

# Print initial values of x and y for t1 and t2
print('t1:', t1.x, t1.y)
print('t2:', t2.x, t2.y)

# Modify the class variable x and the instance variable y for t1
Test.x = 888
t1.y = 999

# Print modified values of x and y for t1 and unchanged values for t2
print('t1:', t1.x, t1.y)
print('t2:', t2.x, t2.y)
````

Note that modifying the class variable x affects all instances of the class, while modifying the instance variable y affects only the specific instance (t1 in this case).

# **Various places to declare static variables:**


1. In general we can declare within the class directly but from out side of any method
2. Inside constructor by using class name
3. Inside instance method by using class name
4. Inside classmethod by using either class name or cls variable
5. Inside static method by using class name

````python
'''mmmmmmmmmmmmmmmmmmmmmmmmmmmmmmm'''
class Test:
    a = 10

    def __init__(self):
        Test.b = 20

    def m1(self):
        Test.c = 30

    @classmethod
    def m2(cls):
        cls.d1 = 40
        Test.d2 = 400

    @staticmethod
    def m3():
        Test.e = 50

# Print the class dictionary before creating an instance
print(Test.__dict__)

# Create an instance of the Test class
t = Test()

# Print the class dictionary after creating an instance
print(Test.__dict__)

# Call the instance method m1
t.m1()

# Print the class dictionary after calling m1
print(Test.__dict__)

# Call the class method m2
Test.m2()

# Print the class dictionary after calling m2
print(Test.__dict__)

# Call the static method m3
Test.m3()

# Print the class dictionary after calling m3
print(Test.__dict__)

# Add a class attribute f
Test.f = 60

# Print the class dictionary after adding attribute f
print(Test.__dict__)
````

````python
class Test:
    """This is the Test class with class attributes and methods."""

    a = 10

    def __init__(self):
        """Initializer method to set class attribute b."""
        Test.b = 20

    def m1(self):
        """Instance method to set class attribute c."""
        Test.c = 30

    @classmethod
    def m2(cls):
        """Class method to set class attributes d1 and d2."""
        cls.d1 = 40
        Test.d2 = 400

    @staticmethod
    def m3():
        """Static method to set class attribute e."""
        Test.e = 50

# Print the docstring of the Test class
print(Test.__doc__)

# Print the class dictionary before creating an instance
print(Test.__dict__)

# Create an instance of the Test class
t = Test()

# Print the class dictionary after creating an instance
print(Test.__dict__)

# Call the instance method m1
t.m1()

# Print the class dictionary after calling m1
print(Test.__dict__)

# Call the class method m2
Test.m2()

# Print the class dictionary after calling m2
print(Test.__dict__)

# Call the static method m3
Test.m3()

# Print the class dictionary after calling m3
print(Test.__dict__)

# Add a class attribute f
Test.f = 60

# Print the class dictionary after adding attribute f
print(Test.__dict__)
````

In this code:

Class variables a, b, c, d2, e, and f are added or modified.
The **\_\_init\_\_** method sets class variable b.
The instance method m1 sets class variable c.
The class method m2 sets class variables d1 and d2.
The static method m3 sets class variable e.
Attribute f is added directly to the class.
The output will show the changes to the class dictionary at each step.

# **How to access static variables:**

1. inside constructor: by using either self or classname
2. inside instance method: by using either self or classname
3. inside class method: by using either cls variable or classname
4. inside static method: by using classname
5. From outside of class: by using either object reference or classname

````python
class Test:
    a = 10

    def __init__(self):
        print(self.a)     # Accessing instance variable 'a' using self
        print(Test.a)     # Accessing class variable 'a' using class name

    def m1(self):
        print(self.a)     # Accessing instance variable 'a' using self
        print(Test.a)     # Accessing class variable 'a' using class name

    @classmethod
    def m2(cls):
        print(cls.a)      # Accessing class variable 'a' using cls
        print(Test.a)     # Accessing class variable 'a' using class name

    @staticmethod
    def m3():
        print(Test.a)     # Accessing class variable 'a' using class name

# Create an instance of the Test class
t = Test()

# Access class variable 'a' using class name
print(Test.a)

# Access instance variable 'a' using instance t
print(t.a)

# Call the instance method m1
t.m1()

# Call the class method m2
t.m2()

# Call the static method m3
t.m3()
````

In a static method (@staticmethod), there's no need to include self or cls as the first parameter. Static methods don't have access to the instance or class itself as the first parameter. They operate on the class-level and don't depend on the instance or class state.

````python
class Test:
    a = 777

    @classmethod
    def m1(cls):
        cls.a = 888

    @staticmethod
    def m2():
        Test.a = 999
        print(Test.a)

# Access class variable 'a' using class name
print(Test.a)

# Call the class method m1
Test.m1()

# Access class variable 'a' using class name
print(Test.a)

# Call the static method m2
Test.m2()

# Access class variable 'a' using class name
print(Test.a)
````

# **How to delete static variables of a class:**

We can delete static variables from anywhere by using the following syntax
del classname.variablename
But inside classmethod we can also use cls variable
del cls.variablename

````python
class Test:
    a = 10

    @classmethod
    def m1(cls):
        del cls.a

# Call the class method m1
Test.m1()

# Print the class dictionary to show the changes
print(Test.__dict__)
````

In this corrected version:

The class method m1 is defined with @classmethod.
Within the m1 method, it uses del cls.a to delete the class variable a.
The method m1 is then called using Test.m1().
Finally, the class dictionary is printed to show the changes.
This code will output an empty dictionary for Test.__dict__ because the class variable a has been deleted inside the class method.

````python
class Test:
    a = 10

    def __init__(self):
        Test.b = 20
        del Test.a

    def m1(self):
        Test.c = 30
        del Test.b

    @classmethod
    def m2(cls):
        cls.d = 40
        del Test.c

    @staticmethod
    def m3():
        Test.e = 50
        del Test.d

# Print the initial class dictionary
print(Test.__dict__)

# Create an instance of Test
t = Test()

# Print the class dictionary after instance creation
print(Test.__dict__)

# Call the method m1
t.m1()

# Print the class dictionary after calling m1
print(Test.__dict__)

# Call the class method m2
Test.m2()

# Print the class dictionary after calling m2
print(Test.__dict__)

# Call the static method m3
Test.m3()

# Print the class dictionary after calling m3
print(Test.__dict__)

# Add a new class variable
Test.f = 60

# Print the class dictionary after adding f
print(Test.__dict__)

# Delete the class variable e
del Test.e

# Print the class dictionary after deleting e
print(Test.__dict__)
````

This code defines a class Test with class variables, instance methods, class methods, and a static method. It demonstrates the manipulation of class variables and their deletion. The __dict__ attribute is used to access the dictionary containing the class's namespace.

# **Static methods**


````python
class MyClass:
    class_variable = "I am a class variable"

    def __init__(self, instance_variable):
        self.instance_variable = instance_variable

    @staticmethod
    def static_method():
        return "This is a static method."

# Calling the static method on the class
result1 = MyClass.static_method()
print(result1)

# Creating an instance of the class
obj = MyClass("I am an instance")

# Calling the static method on the instance (it's less common)
result2 = obj.static_method()
print(result2)
````

In this example, static_method is a static method because it is decorated with @staticmethod. It doesn't have access to the instance or the class variables. It's used when a method doesn't need access to instance-specific or class-specific data.

A static method in Python is a method that belongs to a class rather than an instance of the class. It is defined using the @staticmethod decorator. Unlike regular methods, static methods don't have access to the instance or the class itself (no self or cls parameter), and they can be called on the class without creating an instance.

The main purposes of static methods include:

* Organization: Static methods are used to group functions that logically belong to a class but don't need access to instance-specific data. They provide a way to organize related functionality within a class.

* Namespace: Static methods don't have access to the instance or class variables, so they can't modify the state of an object. They work in a class-level namespace, providing utility functions that are related to the class.

* Code Reusability: Since static methods don't depend on the state of an instance, they can be reused across instances of the class. They are not bound to a particular object's state.

* Class-level Operations: Operations that are not tied to a specific instance but are related to the class as a whole can be implemented as static methods. These methods can be called on the class itself.

Here's an example:

````python
class MathOperations:
    @staticmethod
    def add(x, y):
        return x + y

    @staticmethod
    def multiply(x, y):
        return x * y

# Usage
result_sum = MathOperations.add(3, 5)
result_product = MathOperations.multiply(2, 14)
print(result_sum)
print(result_product)
````

Let's break down the statement "They don't depend on any instance-specific data and can be called on the class directly":

Don't Depend on Instance-Specific Data:

Regular methods in a class (non-static methods) have access to the instance-specific data via the self parameter. This means they can operate on and modify the specific instance's attributes.
In contrast, static methods don't have access to self or any instance variables. They are not bound to a particular instance's state. They are more like standalone functions that are part of the class.
Can Be Called on the Class Directly:

Static methods are associated with the class rather than an instance. You can call them on the class itself, not on an instance of the class.
This means you don't need to create an instance of the class to use a static method. You can call the static method on the class directly, as if it were a standalone function.
Here's a simple example to illustrate:

````python
class MyClass:
    class_variable = 10  # A class variable

    def __init__(self, instance_variable):
        self.instance_variable = instance_variable  # An instance variable

    def regular_method(self):
        return f"Regular method called with instance variable: {self.instance_variable}"

    @staticmethod
    def static_method():
        return "Static method called"

# Creating an instance
obj = MyClass(42)

# Calling a regular method
result_regular = obj.regular_method()
print(result_regular)

# Calling a static method on the class directly
result_static = MyClass.static_method()
print(result_static)
````

In the example, regular_method is a regular method that can access the instance variable, and you need an instance (obj) to call it. On the other hand, static_method is a static method that doesn't depend on the instance-specific data and can be called directly on the class (MyClass).

# **Python Inheritance**





In Python, there are **no keywords like public, protected, or private as explicit access modifiers,** unlike some other object-oriented programming languages. In Python, it follows a philosophy of "we are all consenting adults here," meaning it trusts developers to follow conventions.

Here's what it means in the context of your code:

**If a variable or method name starts with a single underscore (e.g., _variable or _method())**, it is considered a "protected" member. It's a convention, not a strict rule, indicating that it should not be accessed outside the class or subclass, although it is technically still possible.

If a variable or method name starts with **two underscores** (e.g., __variable or __method()), it undergoes name mangling, which makes it harder to access from outside the class. However, it's still accessible with some effort.

If a variable or method name does not start with an underscore, it's considered a "public" member, and it can be freely accessed from outside the class.

So, in Python, you don't have explicit access modifiers, but you have naming conventions that suggest the intended level of access. Developers are expected to follow these conventions to maintain code integrity.

Inheritance is an important aspect of the object-oriented paradigm. Inheritance provides code reusability to the program because we can use an existing class to create a new class instead of creating it from scratch.

In inheritance, the child class acquires the properties and can access all the data members and functions defined in the parent class. A child class can also provide its specific implementation to the functions of the parent class.

In python, a derived class can inherit base class by just mentioning the base in the bracket after the derived class name. Consider the following syntax to inherit a base class into the derived class.


When a child class inherits from a parent class in Python, the memory layout involves the creation of objects and attributes in memory. Let's discuss the key aspects:

* **Class Definition in Memory:**

The parent class and the child class each have their class definitions stored in memory. These definitions include methods, attributes, and other class-level information.
* **Object Creation:**

When you create an instance of the child class (child_obj = Child()), memory is allocated to store the attributes of the object.
The object has attributes inherited from the parent class, and it may have additional attributes defined in the child class.
* **Inherited Attributes:**

The child class object contains attributes inherited from the parent class. These attributes share the memory space with the child object.
In Python, attributes are typically stored in dictionaries, so each object has a dictionary of attributes. When an attribute is accessed, Python looks it up in the dictionary.
* **Overriding and Extending:**

If the child class overrides a method or attribute from the parent class, the child's version is used. If the child extends the functionality, both the parent's and child's versions coexist.
* **Method Resolution Order (MRO):**

Python uses the C3 linearization algorithm to determine the method resolution order, i.e., the order in which it looks for methods in multiple inheritance scenarios.
The MRO influences how inherited methods are accessed and called.
* **Memory Efficiency:**

Python's memory management is designed to be efficient. Shared attributes between parent and child classes are not duplicated in memory for each instance.
Memory is allocated for each object, but common attributes are shared at the class level.
* **Instance and Class Attributes:**

Instance attributes are specific to each object, while class attributes are shared among all instances of a class. Both are stored in memory.

````python
class Parent:
    def __init__(self, x):
        self.x = x

class Child(Parent):
    def __init__(self, x, y):
        super().__init__(x)
        self.y = y

# Create an instance of the Child class
child_obj = Child(20, 30)
print(child_obj.x)
print(child_obj.y)
````

In this example, the memory layout for child_obj might be something like:

````python
child_obj:
{
  'x': 20,  # Attribute from Child
  'y': 30,  # Attribute from Child
  'x': 10   # Attribute from Parent, overwritten by Child
}
````

The super().__init__(x) line in the Child class is used to call the constructor (__init__ method) of the parent class (Parent). This is a way for the child class to initialize the attributes that belong to the parent class before adding its own attributes.

In object-oriented programming, when a class inherits from another class, the child class often needs to perform some additional initialization along with the initialization provided by the parent class. The super() function is used to refer to the parent class, and super().__init__(x) calls the constructor of the parent class, passing the required arguments.

Here's a breakdown:

**Initialization of Parent Attributes:**

The Parent class has a constructor that initializes an attribute x.
When you create an instance of the Child class, you want to make sure that the x attribute of the Parent class is also properly initialized.

**Use of super() Function:**

super() is a built-in function in Python that returns a temporary object of the superclass, which allows you to call its methods.
super().__init__(x) calls the __init__ method of the superclass (in this case, the Parent class) and passes the required argument x to initialize the x attribute of the parent.

**Complete Initialization:**

After the call to super().__init__(x), the Child class can add its own attributes (y in this case) or perform any additional initialization steps specific to the child class.

**Here's the modified example for clarity:**








````python
class Parent:
    def __init__(self, x):
        self.x = x

class Child(Parent):
    def __init__(self, x, y):
        # Call the constructor of the parent class to initialize 'x'
        super().__init__(x)

        # Add additional initialization for the Child class
        self.y = y
````

Without the call to super().__init__(x), the x attribute of the Parent class might not be properly initialized, and the child class could miss out on the initialization logic provided by the parent class. Using super() ensures that both the parent and child classes contribute to the initialization process.

## **Data abstraction in python**


Abstraction is an important aspect of object-oriented programming. In python, we can also perform data hiding by adding the double underscore (___) as a prefix to the attribute which is to be hidden. After this, the attribute will not be visible outside of the class through the object.

Consider the following example.

**Example**

````python
class Employee:
    __count = 0;
    def __init__(self):
        Employee.__count = Employee.__count+1
    def display(self):
        print("The number of employees",Employee.__count)
emp = Employee()
emp2 = Employee()
emp2.display()
try:
    print(emp.__count)
finally:
    emp.display()
````

In Python, a variable or method name starting with a single underscore is considered a "protected" member. While this is more of a convention than a strict rule (Python does not enforce access restrictions like some other languages), it signals to other developers that the variable or method should be treated as a non-public part of the class or module. Here's an example:

````python
class MyClass:
    def __init__(self):
        # A protected variable
        self._protected_variable = 42

    def _protected_method(self):
        print("This is a protected method.")

# Creating an instance of MyClass
obj = MyClass()

# Accessing the protected variable
print(obj._protected_variable)  # Output: 42

# Calling the protected method
obj._protected_method()  # Output: This is a protected method.
````

In this example:

_protected_variable is a protected member of the MyClass class.
_protected_method is a protected method of the MyClass class.
Although it's possible to access and use these members outside the class, the convention of using a single underscore signals that they are intended for internal use and should not be accessed directly by external code. Developers are expected to treat these members with caution and avoid using them inappropriately. However, this is more about convention and communication than strict access control.

In Python, while a variable or method starting with a single underscore is considered a "protected" member, **it is still accessible from outside the class, including by derived classes.** However, the use of a single underscore signals that the member is intended for internal use, and accessing it directly from outside the class is discouraged. Let's illustrate this with an example:

````python
class MyClass:
    def __init__(self):
        # A protected variable
        self._protected_variable = 42

    def _protected_method(self):
        print("This is a protected method.")

class DerivedClass(MyClass):
    def access_protected(self):
        # Trying to access the protected variable from the derived class
        print("Accessing protected variable from DerivedClass:", self._protected_variable)

        # Trying to call the protected method from the derived class
        self._protected_method()

# Creating an instance of the derived class
derived_obj = DerivedClass()

# Accessing the protected variable from the derived class
derived_obj.access_protected()
````

In this example:

MyClass has a protected variable _protected_variable and a protected method _protected_method.
DerivedClass inherits from MyClass.
Even though _protected_variable and _protected_method are marked as "protected," they are still accessible from the DerivedClass because Python does not enforce strict access control. However, it is generally considered good practice to treat these members as if they were private and avoid direct access from outside the class hierarchy.

# **Abstraction in Python**

**Abstract Base Classes**
An abstract base class is the common application program of the interface for a set of subclasses. It can be used by the third-party, which will provide the implementations such as with plugins. It is also beneficial when we work with the large code-base hard to remember all the classes.

**Working of the Abstract Classes**
Unlike the other high-level language, Python doesn't provide the abstract class itself. We need to import the abc module, which provides the base for defining Abstract Base classes (ABC). The ABC works by decorating methods of the base class as abstract. It registers concrete classes as the implementation of the abstract base. We use the @abstractmethod decorator to define an abstract method or ***if we don't provide the definition to the method, it automatically becomes the abstract method.*** Let's understand the following example.

**Example -**



````python
# Python program demonstrate
# abstract base class work
from abc import ABC, abstractmethod

class Car(ABC):
    @abstractmethod
    def mileage(self):
        pass

class Tesla(Car):
    def mileage(self):
        print("The mileage is 30kmph")

class Suzuki(Car):
    def mileage(self):
        print("The mileage is 25kmph")

class Duster(Car):
    def mileage(self):
        print("The mileage is 24kmph")

class Renault(Car):
    def mileage(self):
        print("The mileage is 27kmph")

# Driver code
t = Tesla()
t.mileage()

r = Renault()
r.mileage()

s = Suzuki()
s.mileage()

d = Duster()
d.mileage()
````

An abstract class in Python serves as a blueprint for other classes. It is a class that cannot be instantiated on its own and is meant to be subclassed by other classes. The main purpose of abstract classes is to provide a common interface for all the classes that inherit from it.

Here are some key uses and features of abstract classes:

**Define a Common Interface:**

**Abstract classes define a common interface for all the classes that inherit from them. This helps ensure that all the subclasses adhere to a certain structure and have specific methods.**

**Force Implementation of Methods:**

Abstract classes can include abstract methods, which are methods without a body. Subclasses are required to provide an implementation for these abstract methods. This enforces a certain structure in derived classes.

**Cannot Be Instantiated:**

An abstract class cannot be instantiated on its own. It needs to be subclassed, and an instance of the subclass is created.

**Provide a Template:**

Abstract classes provide a template or a skeleton for derived classes. They define methods that must be implemented in the subclasses but may also provide some default behavior.

**Support Polymorphism:**

Abstract classes support polymorphism. Objects of the derived classes can be used interchangeably with objects of the abstract class type, allowing for flexibility in the design.

**Enforce Design Contracts:**

*`By defining an abstract class with abstract methods, you create a contract that the subclasses must fulfill. This helps enforce a certain design and ensures that specific methods are implemented.`*

Here's a simple example to illustrate the concept:

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius**2

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side**2

# Example usage:
circle = Circle(5)
square = Square(4)

print(circle.area())  # Output: 78.5
print(square.area())  # Output: 16
````

# **Garbage Collection:**


Garbage collection is a process in computer programming languages to automatically reclaim memory occupied by objects that are no longer in use. The primary purpose of garbage collection is to free up memory and prevent memory leaks, where the program retains references to objects that are no longer needed.

In Python, a built-in garbage collector is responsible for automatically managing the memory by reclaiming the space occupied by objects that are no longer reachable or referenced by the program.

Here are some key points about the garbage collector in Python:

**Reference Counting:** Python primarily uses reference counting as a simple garbage collection algorithm. Each object keeps track of the number of references pointing to it, and when this count drops to zero, the memory occupied by the object is freed.

**Cycle Detector:** In addition to reference counting, Python employs a cycle detector to identify and collect objects involved in reference cycles. A reference cycle occurs when a group of objects reference each other, forming a closed loop. Reference counting alone cannot handle such situations, and the cycle detector is responsible for identifying and collecting these cycles.

**gc Module:** Python provides a gc module that exposes the garbage collection features. You can use functions like gc.collect() to manually trigger garbage collection.

Automatic Garbage Collection: In most cases, Python's garbage collector runs automatically in the background, and developers do not need to explicitly manage memory deallocation. The garbage collector identifies and collects objects that are no longer reachable during the execution of the program.

# **Destructors:**


Destructor is a special method and the name should be __del__
Just before destroying an object Garbage Collector always calls destructor to perform clean up
activities (Resource deallocation activities like close database connection etc).
Once destructor execution completed then Garbage Collector automatically destroys that object.
***Note: The job of destructor is not to destroy object and it is just to perform clean up activities.***
Example

````python
import time

class Test:
    def __init__(self):
        print("Object Initialization...")

    def __del__(self):
        print("Fulfilling Last Wish and performing clean up activities...")

# Create an instance of Test
t1 = Test()

# Set t1 to None to delete the reference
t1 = None

# Introduce a delay before the program ends
time.sleep(5)

print("End of application")
````

How to find the number of references of an object:
sys module contains getrefcount() function for this purpose.
he sys.getrefcount() function returns the reference count of an object. It counts how many references are pointing to the object. However, note that the count returned by sys.getrefcount() is always at least 1 more than you might expect because it includes a temporary reference that is created to pass the object to sys.getrefcount().

Here's the explanation:

````python
import sys

class Test:
    pass

t1 = Test()
t2 = t1
t3 = t1
t4 = t1

# The reference count is at least 2 (t1 and the temporary reference used by sys.getrefcount())
print(sys.getrefcount(t1))
````

In this code, t1 is referenced by t2, t3, and t4, so the reference count is 4, but sys.getrefcount() returns 5 because of the temporary reference created by the function itself. The temporary reference is not part of your program logic and is created by the sys.getrefcount() function.
